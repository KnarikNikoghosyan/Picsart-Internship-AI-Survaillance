```python
import cv2
import matplotlib.pyplot as plt
```

```python
img_grayscale = cv2.imread("flower.webp")
```

```python
cv2.imshow("grayscale image", img_grayscale)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Writing an Image

cv2.imwrite("grayscale.jpg", img_grayscale)
```

```python
# Readig an Image

img_color = cv2.imread('flower.webp',cv2.IMREAD_COLOR)
img_grayscale = cv2.imread('flower.webp',cv2.IMREAD_GRAYSCALE)
img_unchanged = cv2.imread('flower.webp',cv2.IMREAD_UNCHANGED)

# or

img_color = cv2.imread('flower.webp',1)
img_grayscale = cv2.imread('flower.webp',0)
img_unchanged = cv2.imread('flower.webp',-1)
```

```python
# Using Matplotlib to display an image

plt.imshow(img_color)
plt.axis("off")
plt.show()
```

```python
plt.imshow(cv2.cvtColor(img_grayscale, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
plt.imshow(cv2.cvtColor(img_unchanged, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
# Displaying an Image using OpenCV

cv2.imshow('color image',img_color)  
cv2.imshow('grayscale image',img_grayscale)
cv2.imshow('unchanged image',img_unchanged)
 
cv2.waitKey(0)
cv2.destroyAllWindows()
```
