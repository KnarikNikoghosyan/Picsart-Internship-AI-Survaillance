```python
import cv2
import numpy as np
```

```python
image = cv2.imread('image.jpg')
```

```python
height, width = image.shape[:2]

# creating translation matrix
tx, ty = width/6, height/6
translation_matrix = np.array([
    [1, 0, tx],
    [0, 1, ty]
], dtype=np.float32)
```

```python
translated_image = cv2.warpAffine(src=image, M=translation_matrix, dsize=(width, height))
```

```python
cv2.imshow('Translated image', translated_image)
cv2.imshow('Original image', image)
cv2.waitKey(0)
cv2.imwrite('translated_image.jpg', translated_image)
```

```python
# Using Matplotlib
import matplotlib.pyplot as plt
#from pylab import rcPerams
#rcPerams['figure.figsize'] = 12, 9
```

```python
image = cv2.imread('image.jpg')
```

```python
height, width = image.shape[:2]
center = (width/2, height/2)
```

```python
plt.imshow(image[:, :, ::-1])
```

```python
rotate_matrix = cv2.getRotationMatrix2D(center=center, angle=45, scale=1)
```

```python
rotated_image = cv2.warpAffine(src=image, M=rotate_matrix, dsize=(width, height))
```

```python
plt.imshow(image[:, :, ::-1])
plt.axis('off')
plt.show()
print('\n')
plt.imshow(rotated_image[:, :, ::-1])
plt.axis('off')
plt.show()
cv2.imwrite('rotated_image2.png', rotated_image)
```

```python
tx, ty = width / 4, height / 4
# creating the translation matrix using tx and ty 
translation_matrix = np.array([
    [1, 0, tx],
    [0, 1, ty]
], dtype=np.float32)
```

```python
translated_image = cv2.warpAffine(
    src=image, M=translation_matrix, dsize=(width, height))
```

```python
plt.imshow(translated_image[:, :, ::-1])
plt.axis('off')
plt.show()
print('\n')
plt.imshow(image[:, :, ::-1])
plt.axis('off')
plt.show()
cv2.imwrite('translated_image2.png', translated_image)
```
